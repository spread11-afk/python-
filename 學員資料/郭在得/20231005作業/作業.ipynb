# 檔案存取
1. 讀取或寫入文件前要先用`open()`開啟文件 -> 產生超連結 -> 可以存取文件。
2. 語法：`file object = open(file_name [, access_mode],encoding='utf-8',newline="")`

    `file_name`：可以是檔案名稱 or 相對位置。
    
    `access_mode`：存取模式，預設為r：僅供閱讀的文件。

    存取模式：

    r：reading 讀取
    
    b：binary 二進位

    w：write (if文件存在：)**覆蓋**;else:建立一個新的文件後寫入。

    a：append (if文件存在：)**添加**;else:建立一個新的文件後寫入。

    

    排列組合：

    |類別|二進位|讀取|寫入|覆蓋|添加|
    |---|---|---|---|---|---|
    |r||O||||
    |r+||O|O|||
    |rb|O|O||||
    |rb+|O|O|O|||
    |w|||O|O||
    |w+||O|O|O||
    |wb|O||O|O||
    |wb+|O|O|O|O||
    |a|||O||O|
    |a+||O|O||O|
    |ab|O||O||O|
    |ab+|O|O|O||O|

    `encoding='utf-8'`：避免亂碼

    `newline`：預設是None，但是設`""`(空字串)比較保險。

## r + read
r是預設其實不用寫，但為了方便觀看下面都會標示。
### read
1. 語法：`read([size])`
2. 讀到該檔案沒有(內容)為止。
3. 以該檔案字元數量讀取時，中/英文/數字/換行都算1格。
4. 若該檔案無參數 -> 打開再關起來 -> 什麼都不會發生
5. `file.closed`：確定檔案有沒有關起來(傳出bool)。

```python
file = open('test1.txt', 'r', encoding='UTF-8') 
content = file.read()
print(content)
file.close( )

print("-------------")

file = open('test1.txt', 'r', encoding='UTF-8') 
content2 = file.read(11)
print(content2)
file.close( )

print("-------------")

file = open('test.txt', 'r', encoding='UTF-8') 
content3 = file.read()
print(content3)
file.close( )

print(file.closed)
```

```python
a="  hello  "
print("---"+a.strip()+"---")
print("---"+a.lstrip()+"---")
print("---"+a.rstrip()+"---")

b="hello, world!"
print(b.startswith("world"))
```

#### 範例：
讀取test1.txt -> 刪除空行和註釋行 -> 按字母順序排列 -> 將結果寫入新創的 result.txt

```python
file=open('test1.txt', 'r', encoding='UTF-8') 
result = list( ) # 建一個空list,儲存資料
# list() -> 把字串轉成字元儲存進list
for line in file.readlines( ):
    line = line.strip( )
    if not len(line) or line.startswith('#'):
    # 檢查是否長度為零（即空行）或以 '#' 開頭（即註釋行）。如果是，則跳過該行，不將其添加到 result 列表中。
        continue 
    result.append(line) # 將line資料加入result
file.close()

result.sort( ) # 按字母順序排列
print(result)
open('result.txt', 'w', encoding='UTF-8').write('%s' % '\n'.join(result))
# %s ：取代str -> 將這個取代成經順序排列的result
# % '\n' ：每行後面用\n分隔
# join ：把多個字串合在一起
```

## 寫入CSV檔
### CSV檔
1. Comma-Separated Values，用逗號分隔的純文字檔 -> 可以用記事本打開(cf. excel)。
2. 資料均沒有類型 -> 都是str。
3. 沒有字體、顏色、儲存格寬度&高度的設置。

### 寫入流程
1. 需先`import csv`
2. 開啟要寫入的CSV檔案，並設定分隔符號(delimiter)和包住str(quotechar)的符號。
3. 使用`writerow()`寫入特定的儲存格。
4. 關閉 CSV 檔案，以確保資料已經寫入並且檔案已儲存。

```python
import csv

with open('test4.csv', 'w', newline='',encoding='utf-8') as file:
    # newline=''：避免多出空行
    csvwriter = csv.writer(file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL) 
    csvwriter.writerow(['編號', '國文', '英文', '數學', '自然', '社會'])
    csvwriter.writerow(['01', 87, 90, 56, 87, 89])
    csvwriter.writerow(['02', 98, 56, 78, 90, 49])
file.close()
```

`quoting=csv.QUOTE_MINIMAL`：以下情況會使用引號
+ 當資料中包含**特殊字元**（如逗號或換行符號）時。
+ 當資料中包含**引號**時，以避免引號與引號之間的字串被誤解為引號的結尾。

### 應用
建立一個test5.txt，內容為：5個隨機50~100的list50組。

```python
import random
import csv

with open('test5.csv','w', newline='', encoding='UTF-8') as file:
    csvWriter = csv.writer(file)
    csvWriter.writerow(['編號', '國文', '英文', '數學', '自然', '社會'])
    for i in range(50):
        scores = [str(i+1)] #把學生編號轉換成str
        for _ in range(5):
            scores += [random.randint(50,100)]
        csvWriter.writerow(scores)
file.close()
```

### 讀取CSV檔
#### reader( )
1. 語法：reader(csvfile, dialect='excel', **fmtparams)

    `csvfile`：.csv的檔。

    `dialect='excel'`：預設為2維的list(像excel一樣，list裡面裝list)
    
    |["a","b","c"]|
    |---|
    |["1","2","3"]|
    |["4","5","6"]|
    
    `**fmtparams`：不限數量的引述呼叫
2. 從0開始，照文件編號輸出完整的**list**。
3. 可以用索引找特定儲存格、內容。

```python
#方法1：設變數
with open('test5.csv','r', encoding='UTF-8') as file:
    a=csv.reader(file)
    List = list(a) # 將讀出來的資料(a)轉成list
print(a.__class__) # 輸出a的變數類型：class csv的reader檔

print(List)

print("----------------")

#方法2：for
with open('test5.csv','r', encoding='UTF-8') as file:
    a=csv.reader(file)
    for i in a:
        print(i)

print("----------------")

#索引1：list
with open('test5.csv','r', encoding='UTF-8') as file:
    a=csv.reader(file)
    list_a=list(a)
print(list_a[2])

print("----------------")

#索引2：設變數
with open('test5.csv','r', encoding='UTF-8') as file:
    a=csv.reader(file)
    for i in a:
        number=i[0]
        chinese=i[1]
        english=i[2]
print(f"編號：{number}：國文：{chinese}分；英文：{english}分")
```

```python
# pop範例
a=[1,2,3,4,5]
a.pop(1)
print(a)

# map範例
b=[1,2,3,4,5]

def double(n):
    return n*2

c=map(double,b)
print(c) # 會印出c的資料型態
print(list(c)) # 把c list才能印出內容
```

## JSON
cf. csv是純文字資料，JSON可以儲存複雜的資料結構。e.g. list裡面有dict，裡面又有list的資料。

優點：

1. 相容性高
2. 格式好閱讀、修改
3. 可以存入許多資料結構(e.g.num, str, bool, null, tuple, list, dict)
4. 很多程式支援JSON(e.g.JS->AJAX->JSON)

### Python cf. JSON的資料型態
|Python|JSON|
|---|---|
|dict|objet|
|list、tuple|array|
|int、float|number|
|True|true|
|False|false|
|None|null|

### Python -> JSON
1. 以 json.dumps( ) 函數從 Python dictionary轉出JSON字串。
2. 以 json.dump( ) 函數從 Python dictionary轉出 JSON 檔案中。
3. json資料於Python處理UTF8碼內容會產生亂碼，建議 dumps 時加入 以下的參數才可以正確處理UTF8碼內容:ensure_ascii=False
4. Python 與 JSON 檔案:
    1. JSON 檔案的中文資料於某些編輯軟體內會變成亂碼，但 Python 可以存取。
    2. 檔案寫入可用一般文件方式寫入，也可以使用 JSON 的方法寫入。

```python
import json
pyDic = {'python': 'good', 'gjun': 100, 'python-class':True,'ICQ': None}
json2 = json.dumps(pyDic, ensure_ascii=False)
print(type(json2))
json3=json2.encode('utf8')
print(type(json3))
print(json2)
print(json3)
```

```python
# 下載json,解析json,轉存為csv檔
import requests
import csv

def download_weather():
    url = 'https://opendata.cwb.gov.tw/fileapi/v1/opendataapi/F-C0032-001?Authorization=rdec-key-123-45678-011121314&format=JSON'

    response = requests.get(url)
    if response.status_code == 200:
        print("下載成功")
        weather = response.json()
        return weather
    else:
        print("下載失敗")
        return False

def parse_json(w):
    location = w['cwbopendata']['dataset']['location']
    weather_list = []
    for item in location:
        city_item = {}
        city_item['城市'] = item['locationName']
        city_item['啟始時間'] = item['weatherElement'][1]['time'][0]['startTime']
        city_item['結束時間'] = item['weatherElement'][1]['time'][0]['endTime']
        city_item['最高溫度'] = float(item['weatherElement'][1]['time'][0]['parameter']['parameterName'])
        city_item['最低溫度'] = float(item['weatherElement'][2]['time'][0]['parameter']['parameterName'])
        city_item['感覺'] = item['weatherElement'][3]['time'][0]['parameter']['parameterName']
        weather_list.append(city_item)
    
    return weather_list

def save_csv(data):
    with open('目前天氣.csv',mode='w',encoding='utf-8',newline='') as file:
        fieldnames = ['城市', '啟始時間','結束時間','最高溫度','最低溫度','感覺']
        writer = csv.DictWriter(file,fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data)

def main():
    #下載json檔
    weather = download_weather()

    if weather != False:
        print("下載完畢")
    else:
        print("應用程式下載失敗")
        return
    
    csv_data = parse_json(weather)
    save_csv(csv_data)


if __name__ == "__main__":
    main()
```

### 檔案進階操作
檔案存取與例外處理

1. Python 的文件存取是透過 open( ) 方法建立文件物件方式進行存取， 如果文件不存在或讀取存限限制將會產生例外。
2. 檔案存取例外為 IOError ，另外有兩個子類別:
    1. FileNotFoundError，代表 “找不到檔案”。
    2. PermissionError，代表 “沒有權限存取”。

## 檔案關閉
1. 檔案開啟後須進行 close( ) 方法進行關閉動作。
2. 若檔案沒關閉會造成:
    1. 開啟的文件物件會占用系統資源。
    2. Python 可以同時間開啟的文件數量有限制 (約 20 份文件)。
    3. 開啟文件物件的模式若為寫入 (w 或 a) 模式，一般都是暫存於緩衝區，系統閒置或文件關閉前才會進行寫入，若沒有進行 close( ) 動作可能造成文 件儲存不完整。

```python
# 操作範例:請動手操作，並留意輸出結果
import os

class tryopenfile:
    def openfile(self, filePath):
        self.handle = open(filePath, 'w')
        self.handle.close()

if __name__ == '__main__':
    t = tryopenfile( )
    filePath = 'test.txt' 
    t.openfile(filePath) 
    os.remove(filePath) 
    print('success')
```

## With 方式操作檔案
1. 多個檔案存取時若前一個檔案產生 IO 例外，將會造成後面無法進行close( ) 動作。
2. 以 with 方式操作檔案:
    1. 檔案存取仍有可能產生 IO 例外。
    2. 離開 with 區塊時檔案將會自動進行close( )動作，自動關閉檔案。

```python
# 操作範例 1:請動手操作，並留意輸出結果
try:
    with open('data.txt', 'r', encoding="utf8") as f:
        print(f.read( ))
except FileNotFoundError:
    print("找不到檔案") 
except PermissionError:
    print("你沒有權限存取")
except IOError: 
    print("其他檔案IO問題")
except: 
    print("其他例外") 
```

```python
import requests
import csv
import io

def download()->list[list] |None:
    # 111年台灣各縣市人口密度
    url="https://data.moi.gov.tw/MoiOD/System/DownloadFile.aspx?DATA=CA18EE06-4A50-4861-9D97-7853353D7108"

    a=requests.request("GET",url)

    try:
        a.raise_for_status()
    except requests.exceptions.RequestException as error:
        print("連線發生錯誤，網路中斷")
        return None
    else:
        if not a.ok:
            print("下載失敗，伺服器錯誤訊息")
            return None
        else:
            file = io.StringIO(a.text)
            a_reader=csv.reader(file)
            next(a_reader)
            return list(a_reader)
```

```python
try:
    a_list=download()
except Exception as i:
    print(f"錯誤：{i}")
else:
    for i in a_list:
        print(i)
```
