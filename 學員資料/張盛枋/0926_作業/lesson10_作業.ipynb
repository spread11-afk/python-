```python
import random
import pyinputplus as pyip

def playDice():
    
    count_same = 0
    comp1 = 0
    comp2 = 0
    comp3 = 0
    comp4 = 0
    comp5 = 0
    comp6 = 0
    result= 0
    Dice_1 = random.randint(1, 6)
    Dice_2 = random.randint(1, 6)
    Dice_3 = random.randint(1, 6)
    Dice_4 = random.randint(1, 6)
    
    print("\n===============骰子遊戲=================:")
    print(f"骰子: {Dice_1}, {Dice_2}, {Dice_3}, {Dice_4}")
        # 骰子比較
    if (Dice_1 == Dice_2):
        comp1 = Dice_1+Dice_2
        count_same+=1
    if (Dice_1 == Dice_3):
        comp2 = Dice_1+Dice_3
        count_same+=1 
    if (Dice_1 == Dice_4):
        comp3 = Dice_1+Dice_4
        count_same+=1
    if (Dice_2 == Dice_3):
        comp4 = Dice_2+Dice_3
        count_same+=1     
    if (Dice_2 == Dice_4):
        comp5 = Dice_2+Dice_4
        count_same+=1
    if (Dice_3 == Dice_4):
        comp6 = Dice_3+Dice_4
        count_same+=1 
    # 骰子比較結果計算
    Dice_sum= Dice_1+Dice_2+Dice_3+Dice_4
    if (count_same == 1):
        result= Dice_sum - comp1 - comp2 - comp3 - comp4 - comp5 - comp6
        print(f"結果:{result}")
    elif(count_same == 2):
        result= comp1
        if (result< comp2):
            result= comp2
        if (result < comp3):
            result= comp3
        if (result < comp4):
            result= comp4
        if (result < comp5):
            result= comp5
        if (result < comp6):
            result= comp6    
        print(f"結果:{result}")   
        
    elif(count_same == 6):
        result= Dice_sum//2
        print(f"結果:{result}")
        
    else:
        print(f"骰子點數不符遊戲規則")
        
while (True):
    playDice()
    answer = pyip.inputYesNo(prompt="還要繼續嗎?(y,n)",yesVal='y',noVal='n')
    print(f"  回覆: {answer}")
    if not (answer == 'y'):
        break
print("遊戲結束")
```
