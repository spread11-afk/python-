```python
import tkinter as tk
import tkintermapview as tkmap  #pip install tkintermapview
```

```python
#-------------------------建立預設柏林地址的地圖---------------------------
#沒有指定center的位置，預設在柏林（52.520006, 13.404954）

class Window(tk.Tk):
   def __init__(self):
       super().__init__()
       #建立地圖，指定地圖的中心位置
       #coner_radius=角半徑，數值越大越柔和，單位是px(但不用寫），大於0->地圖的角落是圓角/等於0->地圖的角落是直角
       map_widget = tkmap.TkinterMapView(self,
                                         width=800,
                                         height=600,
                                         corner_radius=0
                                         )
       map_widget.place(relx=0.5,rely=0.5,anchor=tk.CENTER)
       #place->layout的方法，可以將物件重疊顯示
       #relx,rely 指定物件在視窗的位置(數值為0~1)
       #若要在正中央，必須像上面一樣同時指定xy與tk.center


if __name__ == "__main__":
   window = Window()
   window.geometry("800x600")
   window.title("地圖")
   window.mainloop()
```

```python
#透過座標設定位置
'''
輸入經緯度指定地圖位置
-map_widget.set_position(48.860381, 2.338594) Paris, France

顯示縮放級別，數字越大，地圖就放越大
-map_widget.set_zoom(15) 
'''

class Window(tk.Tk):
    def __init__(self):
        super().__init__()
        #建立地圖
        map_widget = tkmap.TkinterMapView(self,
                                          width=800,
                                          height=600,
                                          corner_radius=0
                                          )
        map_widget.place(relx=0.5, rely=0.5, anchor=tk.CENTER)
        map_widget.set_position(25.038128318756307, 121.56306490172479) #台北市位置
        map_widget.set_zoom(50) #設定顯示大小


if __name__ == "__main__":
    window = Window()
    window.geometry("800x600")
    window.title("地圖")
    window.mainloop()
```
